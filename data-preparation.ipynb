```python
from datetime import datetime
import time
import pandas as pd
import numpy as np
```

```python
output_file_path = 'lowongan_manager_linkedin_scrape.xlsx'
df = pd.read_excel(output_file_path)
print(df.shape)
df.sample(10)
```

### cek duplikasi data

```python
df.duplicated()
```

```python
df[df.duplicated(subset=["Description"], keep=False)] # cek duplicate data berdasarkan kolom Description
df = df.drop_duplicates() # hapus data yang duplicate
```

```python
print(df.shape)
df.sample(10)
```

```python
City = []
Province = []
for row in df['Location']:
    x = row.split(",") #convert each element in the tuple to list
    n_element = len(x) #check how many element in the field (list length)
    #print(n_element) 
    if n_element == 3:
        City.append(x[0]) #take the first element as city
        Province.append(x[1]) #take the second element as province
    elif n_element == 2:
        City.append(np.nan)
        Province.append(x[0])
    else:
        if x[0] == 'Indonesia':
            City.append(np.nan)
            Province.append(np.nan)
        else:
            City.append(x[0])
            Province.append(np.nan)
print('done')
```

```python
#Insert new columns after location. location starts from 0.
df.insert(5, 'City', City)
df.insert(6, 'Province', Province)
print('Done')
```

```python
df.sample(10)
```

```python
fix_list = ['Full-time', 'Part-time', 'Internship', 'Temporary', 'Contract']
new_level = []
new_type = []
length = len(df['Level'])
df_level = df['Level']
df_type = df['Type']
for row in range(length):
    if str(df_level.iloc[row]) in fix_list:
        new_level.append(np.nan)
        new_type.append(df_level.iloc[row]) #iloc = integer location. Function to select row by its index location
    else:
        new_level.append(df_level.iloc[row])
        new_type.append(df_type.iloc[row])
#Insert new columns after location. location starts from 0.
df['Level'] = new_level
df['Type'] = new_type
```

```python
df.head(10)
```

```python
df['Company'].value_counts()
```

```python
df['Title'] = df['Title'].str.lower() #convert all values in Title to lowercase
branch_list= ['business', 'branch', 'development']
scm_list = ['procurement','supply chain', 'logistic', 'warehouse']        
operation_list = ['operation', 'maintenance']
hrga_list = ['hr', 'general affair']
marketing_list =['merchandising', 'sales', 'retail', 'account manager']

length = len(df['Title'])
df_title = df['Title']
title_category = []
for row in range(length):
    if any(element in str(df_title.iloc[row]) for element in branch_list)==True:
        title_category.append('Business Development Manager')
    elif any(element in str(df_title.iloc[row]) for element in scm_list)==True:
        title_category.append('SCM Manager')
    elif any(element in str(df_title.iloc[row]) for element in operation_list)==True:
        title_category.append('Operation Manager')
    elif any(element in str(df_title.iloc[row]) for element in hrga_list)==True:
        title_category.append('HR Manager')
    elif any(element in str(df_title.iloc[row]) for element in marketing_list)==True:
        title_category.append('Marketing Manager')
    else:
        title_category.append('Others')

#Insert new columns after location. location starts from 0.
df['Title Category'] = title_category
```

```python
# Slice only job lists with title category = Marketing Manager
df_ds = df[df['Title Category']=='Marketing Manager']
```
